```python
import pandas as pd

from fast_ml.utilities import display_all
from fast_ml.feature_selection import get_constant_features
```

```python
df = pd.read_csv("dataset_1.csv")
df.head()
```

```python
#what this will do is get a constant features, we are also setting a threshold to drop values that are 99% constant
constant_features = get_constant_features(df, threshold = 0.99, dropna = False)

#the output is a dataframe with description, name of the variable, value, and how 
constant_features.head()
```

```python
constant_features_list = constant_features.query("Desc== 'Constant' ")['Var'].to_list()
print(constant_features_list)
```

```python
#the quasiconstant features are also in the constant feautures dataframe so now we want to query for it
quasi_constant_list = constant_features.query("Desc == 'Quasi Constant' ")['Var'].to_list()
print(quasi_constant_list)
```

```python
df.drop(columns = constant_features_list, inplace = True)
df
```

```python
df.drop(columns = quasi_constant_list, inplace = True)
df
```
