```python
import sympy as sp
```

```python
x, y, z, alpha, beta = sp.symbols('x y z alpha beta', real = True)
i, j, k  = sp.symbols('i j k', integer = True, positive = True)
```

```python
sp.plot(x ** 2, show = True)
```

```python
sp.plot(x ** 2, (x, -10, 10), tite = "x^2", show = True)
```

```python
sp.plot(x ** 2 - 5, x ** 2 + 5, x ** 2, (x, -10, 10), tite = "x^2", show = True)
```

```python
p = sp.plot(x ** 2, (x, -10, 10), show = False)
p.extend(sp.plot(x ** 2 - 5, (x, -10, 10 ), show = False))
p.show()
```
