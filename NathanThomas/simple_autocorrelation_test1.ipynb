this is for making autocorrelation of a time series

```python
import numpy as np 
```

```python
#first make a list
x = [0]
```

```python
#build a loop
for i in range(1000):
    
    #what this loop is doing is that it is adding a random number, and then we are going to add the random number to the old number with an arbitrary factor (we used 0.3) 
    x.append(np.random.randn() + x[-1]*0.3)

    #what we expect is a positive correlation at the value of t to the value of t-1
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.plot(np.cumsum(x))
```

```python
#by using np.cumsum then you are making it a time series because they will all connect
```

```python
#we plot all of the values at time t by t-1
plt.plot(x[:-1], x[1:], 'o')
```

```python
#there doesn't seem to be that much autocorrelation so instead we will change that factor up to show more autocorrelation

y = [0]

for j in range(1000):
    
    y.append(np.random.randn() + y[-1]*0.9)

plt.plot(y[:-1], y[1:], 'o')
```

```python
#there is more autocorrelation in this let's also look at the time series of this

plt.plot(np.cumsum(y))
```

```python
#instead of adding the previous value we subtract the previous value

z = [0]

for l in range(1000):

    z.append(np.random.randn() - z[-1] * 0.9)
```

```python
plt.plot(np.cumsum(z))
#it is a lot more mean reverting, like it jumps back to the main value
```

```python
plt.plot(z[:-1], z[1:], 'o')
```

```python
#now we want to check to see if the autocorrelation is related to the t-1 and t-2, for that we will have to initialize two variables

a = [0,0]

for m in range(1000):

    a.append(np.random.randn() + a[-1]*0.9 + a[-2]*0.3)

#by making the second constant for the t-2 spot we are saying that time has less of an influence on the market
```

```python
plt.plot(np.cumsum(a))
```

```python
#the values we had picked were two strong so we'll have to do it again with weaker values

b = [0,0]

for n in range(1000):

    b.append(np.random.randn() + b[-1]*0.01 + b[-2]*0.3)

plt.plot(np.cumsum(b))
```

```python
plt.plot(b[:-1], b[1:], 'o')
```

```python
plt.plot(b[:-2], b[2:], 'o')
```
