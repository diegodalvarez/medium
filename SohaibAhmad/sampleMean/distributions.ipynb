```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
#this is random normally distributed 
df = pd.DataFrame(np.random.randn(10000,1))
```

```python
#this finds the mean and the standard deviation
pop_mu = df.mean(axis = 0)
pop_st = df.std(axis = 0)
```

```python
#this generates sample means and standard deviations
s_mu = [df.sample(100).mean()[0] for i in range(1000)]
```

```python
plt.figure(figsize = (20,10))
sns.distplot(s_mu).grid()
plt.title('Sampling Distribution of Sample Mean (100 samples were N = 1,000')
plt.axvline(x = np.mean(s_mu), label = "mean of sample means")
plt.axvline(x = np.mean(s_mu) + np.std(s_mu), label = "std of sample means", color = 'r')
plt.axvline(x = np.mean(s_mu) - np.std(s_mu), label = "std of sample means", color = 'r')
plt.legend()
plt.show()
```
