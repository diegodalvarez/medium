```python
#we want to plot sepal_length, sepal_wdith, petal_length, and petal_width all in one plot first we show make our variables x, y, z, c

#this makes the figure
fig = plt.figure(figsize = (15,8))

#this makes the subplot (the thing that has the index on it). The 111 saying you want it on the first of each, not sure what the projection does
ax = fig.add_subplot(111, projection = '3d')

#then we make all of the variables
x = iris['sepal_length']
y = iris['sepal_width']
z = iris['petal_length']
c = iris['petal_width']

#this looks like it makes the scatter plot although it am not sure what the cmap = plt.hot means, but I think it has to do something with type of color for the 4d
img = ax.scatter(x, y, z, c = c, cmap = plt.hot())

#this adds in the necessary colorbar for the heat map
fig.colorbar(img)

#this plots it
plt.show()
```

```python
#okay lets try to make two different data data sets. Then we will plot both and map them differently 
iris.tail()
```

```python
#the first one will be (x, y, z) will be (sepal_length, sepal_width, petal_length) with color being petal_width
#the second one will be (x, y, z) will be (sepal_width, petal_length, petal_width) with color being sepal_length
#we will map the first one in hot and the second one in cool
```

```python
#let's fist make points for each plot

x1, y1, z1, c1 = iris['sepal_length'], iris['sepal_width'], iris['petal_length'], iris['petal_width']
x2, y2, z2, c2 = iris['sepal_width'], iris['petal_length'], iris['petal_width'], iris['sepal_length'] 
```

```python
#now lets plot them 

fig = plt.figure(figsize = (15,7))
ax = fig.add_subplot(111, projection = '3d')
img1 = ax.scatter(x1, y1, z1, c = c1, cmap = plt.hot())
img2 = ax.scatter(x2, y2, z2, c = c2, cmap = plt.cool())
fig.colorbar(img1)
fig.colorbar(img2)
plt.show()
```

```python
#you can plot in 3d but it is pretty pointless, this would look like the bubbles of each point being a different size, because the iris dataset is only so big we will just use the np.random for size

fig = plt.figure(figsize = (15,8))
ax = fig.add_subplot(111, projection = '3d')

x = iris['sepal_length']
y = iris['sepal_width']
z = iris['petal_length']
c = iris['petal_width']
size = (np.random.randn(len(c)) + 1) ** 10

img = ax.scatter(x, y, z, c = c, cmap = plt.hot(), s = size)
fig.colorbar(img)
plt.show()
```
