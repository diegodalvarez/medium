```python
import numpy as np 
import pylab as plt
```

```python
#number of days
n_days = 252

#number of monte carlo simulations we want to run
n_runs = 100000

#starting price of the underlying stock
spot_price = 100

#the strike price of the stock
strike = 100

#annualized implied volatility
volatility = 0.15
```

```python
#then they create an array of parameters to it will be a 100,000 x 252 normally distributed random numbers representing daily percentage retursn of the underlying security

#this gets the random numbers
np.random.seed(25)

rets = np.random.randn(n_runs, n_days) * volatility / np.sqrt(252)
rets.shape
```

```python
traces = np.cumprod(1 + rets, 1) * spot_price

for i in traces[:100,:]:
    plt.plot(i)

plt.grid()
plt.xlabel('days', fontsize = 12)
plt.ylabel('Spot Price', fontsize = 12)
plt.show()

```

```python
plt.hist(traces[:-1], bins = 40);
plt.title('Distribution of final prices')
plt.xlabel('Final prices', fontsize = 12)
plt.show()
```

```python
call = np.mean((traces[:, -1] - strike) * ((traces[:, -1] - strike) > 0))
call
```
